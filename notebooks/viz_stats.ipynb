```python
import torch
import plotly.express as px

from datasets import _get_train_stats
```

```python
stats = _get_train_stats("train")
```

```python
px.imshow(
    stats["x_mean"],
    facet_col=0,
    color_continuous_scale="Rainbow",
)
```

```python
px.imshow(
    stats["x_std"],
    facet_col=0,
    color_continuous_scale="Rainbow",
)
```

```python
px.imshow(
    stats["x_std"] == 0,
    facet_col=0,
    color_continuous_scale="Rainbow",
)
```
